```python
import pandas as pd
import numpy
import nltk
import os
import re
```

```python
import docx2txt
import re

# Load the Word document
doc_file = "5 papers related to blockchain.docx"
doc_text = docx2txt.process(doc_file)

# Save the text as a .txt file
with open("blockchain.txt", "w", encoding="utf-8") as f:
    f.write(doc_text)

# Read the .txt file and clean the text
with open("blockchain.txt", "r", encoding="utf-8") as f:
    text = f.read()

clean_text = re.sub(r'\n', ' ', text)    # Remove newlines
clean_text = re.sub(r'\d+', '', clean_text)    # Remove numbers
clean_text = re.sub(r'[^\w\s]', '', clean_text)    # Remove non-alphanumeric characters
clean_text = re.sub(r'\s+', ' ', clean_text)    # Remove spaces
clean_text = clean_text.lower()    # Convert to lowercase

# Save the cleaned text as a list
lis= clean_text.split(' ')

#print(clean_text_list)

```

```python
#Generating Unigram
```

```python
tokens = [t for l in lis for t in l.split() ] 
```

```python
freq = nltk.FreqDist(tokens)
```

```python
freq
```

```python
tl = len(tokens)
sum(freq.values())
```

```python
probs = {k: v/tl for (k,v) in freq.items()}
```

```python
from operator import itemgetter
sprobs = dict(sorted(probs.items(), key=itemgetter(1), reverse=True))
```

```python
pvals = list(sprobs.values())
cumprobs = {k: sum(pvals[0:ix+1]) for ix, (k,v) in enumerate(sprobs.items())}
```

```python
import random
random.seed(42)
random.random()
```

```python
keys = list(cumprobs.keys())
values = cumprobs.values()
for i in range(10):
    rn = random.random();
    #print(rn)
    ix = [i for (i, v) in enumerate(values) if v > rn][0]
    #print(rn, ix, keys[ix])
    print(keys[ix])
```

```python
#Generating bigram
```

```python
import nltk
```

```python
bigrams=list(nltk.bigrams(tokens))
```

```python
bigrams
```

```python
freq = nltk.FreqDist(bigrams)
```

```python
freq
```

```python
tl = len(bigrams)
sum(freq.values())
```

```python
probs = {k: v/tl for (k,v) in freq.items()}
from operator import itemgetter
sprobs = dict(sorted(probs.items(), key=itemgetter(1), reverse=True))
pvals = list(sprobs.values())
cumprobs = {k: sum(pvals[0:ix+1]) for ix, (k,v) in enumerate(sprobs.items())}
import random
random.seed(42)
random.random()
keys = list(cumprobs.keys())
values = cumprobs.values()
for i in range(10):
    rn = random.random();
    #print(rn)
    ix = [i for (i, v) in enumerate(values) if v > rn][0]
    #print(rn, ix, keys[ix])
    print(keys[ix])
```

```python
for i in range(5):
    # Generate sentence using bigram model
    sentence = " "
    token = bigrams
    for j in range(10):
        choices = [bigram[1] for bigram in freq if bigram[0] == token]
        if choices:
            token = random.choice(choices)
        else:
            token = random.choice(tokens)
        sentence += token + " "
    print("Bigram sentence:", sentence.strip())
```

```python
sentence = " "
token = trigrams
for j in range(10):
    choices = [trigram[2] for trigram in freq if trigram[0] == token[0] and trigram[1] == token[1]]
    if choices:
        token = (token[1], random.choice(choices))
    else:
        token = (" ", random.choice(tokens))
        sentence += token[1] + " "
print("Trigram sentence:", sentence.strip())
```

```python
for i in range(5):
    sentence = " "
    token =Fourgram
    for j in range(10):
        choices = [fourgram[3] for fourgram in freq if fourgram[0] == token[0] and fourgram[1] == token[1] and fourgram[2] == token[2]]
        if choices:
            token = (token[1], token[2], random.choice(choices))
        else:
            token = (" ", "  ", random.choice(tokens))
            sentence += token[2] + " "
    print("4-gram sentence:", sentence.strip())
```
